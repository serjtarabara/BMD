```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional metod
plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sample plot')

```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'g')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'y')
```

```python
#oop metod
fg = plt.figure()

axes_1 = fg.add_axes([0, 0, 1, 1])

axes_1.plot(x, y)
axes_1.set_xlabel('X axis')
axes_1.set_ylabel('Y axis')
axes_1.set_title('00 plot')
```

```python
fg = plt.figure()

axes_1 = fg.add_axes([0, 0, 1, 1])
axes_2 = fg.add_axes([0.1, 0.5, 0.4, 0.3])

axes_1.plot(x, y, 'b')
axes_1.set_xlabel('X axis')
axes_1.set_ylabel('Y axis')
axes_1.set_title('00 1 plot')

axes_2.plot(x, y, 'r')
axes_2.set_xlabel('X axis')
axes_2.set_ylabel('Y label')
axes_2.set_title('00 2 plot')
```

```python
fg, axes = plt.subplots()

axes.plot(x, y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('00 plot')
```

```python
fg, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y)
axes[1].plot(y, x)
```

```python
#Debug
fg, axes = plt.subplots(nrows = 4, ncols = 4)
plt.tight_layout()
```

```python
fg = plt.figure()

axes = fg.add_axes([0, 0, 1, 1])

axes.plot(x, x**2, label = 'x^2')
axes.plot(x, x**3, label = 'x^3')
axes.legend()
```

```python
#Size and DPI change
fg = plt.figure(figsize = (8, 4), dpi = 100)
```

```python
fg, axes = plt.subplots()
axes.plot(x, y, lw = 5, ls = '--')
```

```python
fg, axes = plt.subplots()
axes.plot(x, y, marker = 'o', 
          markersize = 20, 
          markerfacecolor = 'r')
```
