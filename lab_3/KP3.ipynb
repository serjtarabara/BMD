```python
import seaborn as sb
```

```python
%matplotlib inline
```

```python
dots = sb.load_dataset('dots')
```

```python
dots.head()
```

```python
sb.distplot(dots['firing_rate'])
```

```python
sb.jointplot(x = 'firing_rate', y = 'coherence', data = dots, kind = 'hex')
```

```python
sb.pairplot(dots, palette = sb.color_palette("Paired"), hue = 'choice')
```

```python
sb.barplot(x = 'choice', y = 'firing_rate', data = dots)
```

```python
import numpy as np
sb.barplot(x = 'choice', y = 'firing_rate', data = dots, estimator = np.median)
```

```python
sb.boxplot(x = "coherence", y = "firing_rate", data = dots, palette = 'rainbow')
```

```python
sb.stripplot(x = "coherence", y = "firing_rate",data = dots)
```

```python
sb.heatmap(dots.corr())
```

```python
import matplotlib.pyplot as plt
g = sb.FacetGrid(dots, col = "coherence", row = "choice")
g = g.map(plt.hist, "firing_rate")
```
